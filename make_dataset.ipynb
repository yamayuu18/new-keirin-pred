```python
from __future__ import annotations

import glob
import pandas as pd
import pickle
from natsort import natsorted
from sklearn.preprocessing import LabelEncoder
from typing import Final
```

```python
pd.get_option('display.max_columns')
pd.set_option('display.max_columns', 450)
```

```python
DROP_COL: Final[list[str]] = [
    'レースグループ名', '選手名'
]
```

```python
def get_df() -> pd.DataFrame:
    files = natsorted(glob.glob('race_data/race_data*.csv'))

    race_data_list: pd.DataFrame = []
    for file in files:
        try:
            race_data_list.append(pd.read_csv(file, encoding='utf-8', header=0))
        except UnicodeDecodeError:
            race_data_list.append(pd.read_csv(file, encoding='cp932', header=0))
    df_race_data = pd.concat(race_data_list)
    
    return df_race_data
```

```python
df_race_data = get_df()
```

```python
# 開催日時のdatetime型への変換
df_race_data['開催日時'] = pd.to_datetime(df_race_data['開催日時'])
```

```python
# 重複行の削除
df_race_data.drop_duplicates(subset=['開催日時', '開催場所', '車番'], inplace=True)
```

```python
df_race_data.head(1)
```

```python
#  object型の抽出
df_race_data.select_dtypes(include=object).head(1)
```

```python
# int型の抽出
df_race_data.select_dtypes(include=int).head(1)
```

```python
# int型の欠損値埋め
for col in df_race_data.select_dtypes(include=int):
    df_race_data[col].fillna(0, inplace=True)
```

```python
# float型の抽出
df_race_data.select_dtypes(include=float).head(1)
```

```python
# float型の欠損値埋め
for col in df_race_data.select_dtypes(include=float):
    df_race_data[col].fillna(0.0, inplace=True)
```

```python
# レース名が入っている列のobject型への変換
for col in df_race_data.filter(like='レース名'):
    df_race_data[col] = df_race_data[col].astype('object')
```

```python
df_race_data.select_dtypes(include=float).head(1)
```

```python
# 着を含む列をint型に変換
for col in df_race_data.select_dtypes(include=object).filter(like='着'):
    df_race_data[col].replace(['棄', '失', '欠'],[10, 10, 10], inplace=True)
    df_race_data[col].fillna(0, inplace=True)
    df_race_data[col] = df_race_data[col].astype('int')
```

```python
df_race_data.select_dtypes(include='object')
```

```python
# DROP_COL列の削除
df_race_data.drop(DROP_COL, axis=1, inplace=True)
```

```python
df_race_data.head(1)
```

```python
# object型の列の欠損をnullに置換
for col in df_race_data.select_dtypes(include=object):
    df_race_data[col].fillna('Null', inplace=True)
```

```python
df_race_data.head(1)
```

```python
# LabelEncoderをかけるために、object型のdataframeを作成
df_no_order = df_race_data.select_dtypes(include=object)
df_no_order.head(1)
```

```python
categorical_dims = {}
```

```python
# レース名のLabelEncoder
le = LabelEncoder()
df_race_data['レース名'] = le.fit_transform(df_no_order['レース名'])
df_no_order.drop('レース名', axis=1, inplace=True)
categorical_dims['レース名'] = len(le.classes_)
```

```python
def labelencoder(col: str) -> None:
    unique_col = []
    df_temp = df_no_order.filter(like=col)
    for col in df_temp:
        for unique_d in df_race_data[col].unique():
            unique_col.append(unique_d)
    unique_col = list(set(unique_col))
    le.fit(unique_col)
    for col in df_temp:
        df_race_data[col] = le.transform(df_race_data[col])
        df_no_order.drop(col, axis=1, inplace=True)
        categorical_dims[col] = len(le.classes_)
```

```python
# その他のレース名のLabelEncoder
labelencoder('レース名')
```

```python
# それ以外のLabelEncoder
for col in df_no_order.columns:
    df_race_data[col] = le.fit_transform(df_race_data[col])
    df_no_order.drop(col, axis=1, inplace=True)
    categorical_dims[col] = len(le.classes_)
```

```python
# 試行(グループごとにデータフレームを抽出)
for race, sub_df in df_race_data.groupby(['開催日時', '開催場所']):
    display(race)
    display(sub_df)
    break
```

```python
#  グループ単位にID付与
df_race_data['グループID'] = df_race_data.groupby(['開催日時', '開催場所']).grouper.group_info[0]
```
